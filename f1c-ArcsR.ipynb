```r

library(ggraph)
library(tibble)

library(tidygraph)
```

```r

pdf(file = "../plots/Arcs_H3K27ac.pdf",   # The directory you want to save the file in
    width = 5, # The width of the plot in inches
    height = 5)


t = tibble(
    from=c('promoter1', 'promoter2', 'enhancer1'),#, 'promoter1', 'promoter2', 'enhancer1'),
    to=c('promoter2','enhancer1','enhancer2'),#,'promoter2','enhancer2','enhancer2')
    w=c(16.750905924984703,36.65772506941503, 46.591369005600264),
    
)

ggraph(t, layout="linear") +
    geom_edge_arc(edge_colour="#01509f", edge_alpha=0.5, aes(width = w), fold=TRUE) + 
    theme_void() +
    theme(
        plot.title = element_text(hjust = 0.5, size=30),
        legend.position="none",
        plot.margin=unit(c(0,0,0.4,0), "null"),
        panel.spacing=unit(c(0,0,3.4,0), "null")
    ) +
    geom_node_point(size=10, color=c('#000000', '#000000','#000000','#000000')) +
    geom_node_text(aes(label=c('Promoter', 'Promoter','Enhancer','Enhancer')), angle=90, hjust=1, nudge_y = -0.1, size=5) +
    ylim(-0.5,0.5) +
    ggtitle('H3K27ac')



dev.off()



pdf(file = "../plots/Arcs_H3K4me3.pdf",   # The directory you want to save the file in
    width = 5, # The width of the plot in inches
    height = 5)


t = tibble(
    from=c('promoter1', 'promoter2', 'enhancer1'),#, 'promoter1', 'promoter2', 'enhancer1'),
    to=c('promoter2','enhancer1','enhancer2'),#,'promoter2','enhancer2','enhancer2')
    w=c(25.249423697556477,45.48271092669433, 29.267865375749192),
    
)

ggraph(t, layout="linear") +
    geom_edge_arc(edge_colour="#098e37", edge_alpha=0.5, aes(width = w), fold=TRUE) + 
    theme_void() +
    theme(
        plot.title = element_text(hjust = 0.5, size=30),
        legend.position="none",
        plot.margin=unit(c(0,0,0.4,0), "null"),
        panel.spacing=unit(c(0,0,3.4,0), "null")
    ) +
    geom_node_point(size=10, color=c('#000000', '#000000','#000000','#000000')) +
    geom_node_text(aes(label=c('Promoter', 'Promoter','Enhancer','Enhancer')), angle=90, hjust=1, nudge_y = -0.1, size=5) +
    ylim(-0.5,0.5) +
    ggtitle('H3K4me3') + scale_x_reverse()


dev.off()
```
